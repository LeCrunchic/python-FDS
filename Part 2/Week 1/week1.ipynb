```
# Visualize the shape of the population data by making a histogram.

sns.distplot(data['name_letters'])
```

```
# True mean and std of the population

truMean = data['name_letters'].mean()
truStd = data['name_letters'].std()

print('True Mean:', truMean)
print('True Standard Deviation:', truStd)
```

```
# Draw 1,000 samples of size n=5 from the population data.  Calculate the mean of each sample. 

thousandSamples5 = np.zeros((1000, 1))
np.random.seed(87)

for i in range(1000):
    thousandSamples5[i,:] = np.random.choice(data['name_letters'].values, size=5).mean()    
```

```
# Graph these 1,000 sample means in a histogram and examine the shape.
sns.distplot(thousandSamples5)
```

```
# Calculate the mean and standard deviation of the sampling distribution.
mean5 = thousandSamples5.mean()
std5 = thousandSamples5.std()

print('With n=5')
print('Mean of sampling distribution', mean5)
print('Standard Deviation of sampling distribution:', std5)
```

```
# Same process as before but with n=15

thousandSamples15 = np.zeros((1000, 1))

for i in range(1000):
    thousandSamples15[i,:] = np.random.choice(data['name_letters'], size=15).mean()
    
sns.distplot(thousandSamples15)

mean15 = thousandSamples15.mean()
std15 = thousandSamples15.std()

print('with n=15')
print('Mean of sampling distribution:', mean15)
print('std of sampling distribution:', std15)
```

```
# with n=25

thousandSamples25 = np.zeros((1000, 1))

for i in range(1000):
    thousandSamples25[i,:] = np.random.choice(data['name_letters'], size=25).mean()
    
sns.distplot(thousandSamples25)

mean25 = thousandSamples25.mean()
std25 = thousandSamples25.std()

print('with n=25')
print('mean of sampling distribution:', mean25)
print('std of sampling distribution:', std25)
```
